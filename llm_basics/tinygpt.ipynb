```python
# Code for buulding a tiny GPT model

import pandas as pd
import numpy as np
import torch
```

```python
def encoder(text):
    encoded = []
    for ch in text:
        encoded.append(stoi[ch])
    return encoded

def decoder(encoded):
    text = ''
    for i in encoded:
        text += itos[i]
    return text

with open ('input.txt', 'r', encoding='utf-8') as file:
    text = file.read().strip().replace('\n', ' ')

char_set = sorted(list(set(text)))
vocab_size = len(char_set)
#print(''.join(char_set), ', Size:' , vocab_size)

stoi = {} # Char to Int
itos = {} # Int to Char

for i, ch in enumerate(char_set):
    stoi[ch], itos[i] = i, ch


data = torch.tensor(encoder(text), dtype=torch.long)
#print(data.shape, data.dtype)

n = int(0.9*len(data))
train_data = data[:n]
val_data = data[n:]
```
